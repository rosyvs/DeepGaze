**Local Regressions**:
1 if the immediate previous fixation is a higher IA_ID, 0 if not,
and NaN if either the previous or current IA is NaN (in other words, the fixation was not in an interest
area

**Labeling**
Can determine the sequence that the label matches by looking at the CURRENT_FIX_START and CURRENT_FIX_END. Label is given by LOCAL_REGRESSION 

**Prediction**
To predict if a timepoint in a sequence is a local regression the model needs to learn to look at the previous fixation and see whether the current point is during a fixation that has a lower IA_ID (meaning that the current fixation is on a word that occurred before the previous fixations word)
